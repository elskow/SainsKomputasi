```python
import newton_raphson as nr
import secant as sc
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fun = "10*x**3 + 5*x**2 + 6*x - 3717"
```

```python
init_x1 = 1
init_x2 = 2
```

```python
_n_iter = 17
```

```python
newton = nr.NewtonRaphson(fun, init_x1)
secant = sc.Secant(fun, init_x1, init_x2)
```

```python
newton.solve(_n_iter)
```

```python
secant.solve(_n_iter)
```

```python
history_newton = newton.get_iterations()
history_secant = secant.get_iterations()
```

```python
error_newton = [history_newton[i]["error"] for i in range(1, len(history_newton) + 1)]
error_secant = [history_secant[i]["error"] for i in range(1, len(history_secant) + 1)]

plt.style.use("seaborn")
sns.set_style("white")

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(error_newton, color="red", label="Newton-Raphson", marker="o", linestyle="--", linewidth=1.5, markersize=5,
        markerfacecolor="red", markeredgecolor="red", markeredgewidth=1.5, alpha=0.7, zorder=2, clip_on=False)
ax.plot(error_secant, color="green", label="Secant", marker="o", linestyle="-.", linewidth=1.5, markersize=5,
        markerfacecolor="green", markeredgecolor="green", markeredgewidth=1.5, alpha=0.7, zorder=2, clip_on=False)
ax.set_xlabel("Iterations")
ax.set_ylabel("Error")
ax.set_xticks(range(0, _n_iter))
ax.set_xticklabels(range(1, _n_iter + 1))
ax.set_title("Performance of Newton-Raphson and Secant methods\n(Symlog scaled)")
ax.grid(color="grey", linestyle="-.", linewidth=0.5, alpha=0.4, zorder=1, clip_on=False, axis="y")

lowest_error_newton = min(error_newton)
lowest_error_secant = min(error_secant)

plt.text(0.5, 0.5,
         f"Lowest error for Newton-Raphson: {lowest_error_newton:.2e}\nLowest error for Secant: {lowest_error_secant:.2e}",
         horizontalalignment="center", verticalalignment="center", transform=ax.transAxes, fontsize=10, color="black",
         alpha=0.3, zorder=1, clip_on=False,
         bbox=dict(boxstyle="square", facecolor="white", edgecolor="black", alpha=0.3, pad=0.6))

ax.legend()
ax.set_yscale("symlog")
ax.set_ylim(bottom=0)
plt.show()
```
